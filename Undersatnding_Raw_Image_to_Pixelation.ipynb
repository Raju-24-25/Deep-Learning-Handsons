## Understanding the `Raw Image` to `Pixelation`

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from PIL import Image
```

```python
image_path = r"/content/zero.png"
```

```python
img = Image.open(image_path)
```

```python
img
```

##Convert the `Raw digital image` to `Std. RGB image`

```python
img = img.convert('RGB')
img
```

# Convert the above image to `Grayscale`

```python
img_gray = img.convert('L')
img_gray
```

```python
width, height = img_gray.size
print('The Height of the Number 0 is :', height, 'px')
print('The Width of the Number 0 is :', width, 'px')
```

```python
img_gray_resized = img_gray.resize((28, 28))
img_gray_resized
```

```python
width1, height1 = img_gray_resized.size
print('The Height of the Number 0 is :', height1, 'px')
print('The Width of the Number 0 is :', width1, 'px')
```

```python
final_img_array = np.array(img_gray_resized)
final_img_array
```

```python
print('Dimension of Pixelated Array :', final_img_array.ndim)
print('Shape of Pixelated Array :', final_img_array.shape)
```

## `Plotting`

```python
plt.figure(figsize=(4,4))
plt.imshow(final_img_array, cmap='gray')
plt.colorbar()
plt.show()
```
